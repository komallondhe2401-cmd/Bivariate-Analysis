# Scatterplot

```python
#relationship between total bill and tip in tips dataset

sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker')
```

```python
#Barplot
sns.barplot(data=titanic,x='Pclass',y='Age',hue='Sex')
```

```python
sns.barplot(data=titanic,x='Pclass',y='Fare')
```

```python
#Boxplot
sns.boxplot(data=titanic,x='Sex',y='Age',hue='Survived')
```

```python
#Distplot
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
#HeatMap
titanic.head(5)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
titanic.groupby('Sex')['Survived'].mean()*100
```

```python
titanic.groupby('Embarked')['Survived'].mean()*100
```

```python
sns.heatmap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

```python
#Cluster Map
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
